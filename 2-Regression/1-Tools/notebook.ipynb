# Welcome to your notebook

```python
import matplotlib.pyplot as plt  # Grafik çizimi için gerekli kütüphane
import numpy as np  # sayısal işlemler için gerekli kütüphane
from sklearn import datasets, linear_model, metrics, model_selection  # makine öğrenme için gerekli kütüphaneler

# Dataset yükleme
# X satırları, y sütunları
# X: Özellikler (features), y: Etiketler (labels)
X, y = datasets.load_diabetes(return_X_y=True)

# Veri setinin boyutlarını yazdır
print(X.shape)  # veri boyutları
print(X[0]) # İlk örnek verisi
```

```python
# 2. indeksteki sütunu çıkar
X = X[:, 2] # İkinci sütunu al
print(X.shape)
```

```python
# Veriyi 2D array (dizi) haline getir
X = X.reshape((-1, 1)) # -1 satır sayısını otomatik olarak belirler, 1 sütun sayısı
print(X.shape)
```

```python
# Veriyi eğitim ve test verilerine ayır
# veri kümesini eğitim ve test verilerine ayırır
# test_size=0.33, verinin %33'ü test için kullanılacak
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, y,
    test_size=0.33,
    random_state=42  # random_state, her seferinde aynı bölmeyi elde etmek için kullanılır
)
```

```python
# Lineer regresyon modeli oluştur ve verilerimizle eğit
model = linear_model.LinearRegression() # Lineer regresyon modeli oluştur
model.fit(X_train, y_train) # Modeli eğitim verileri ile eğit
```

```python
# Test verilerimizi kullanarak tahmin yap
y_pred = model.predict(X_test) # Test verileri ile tahmin yap
```

```python
# y = w * x + s
print(model.coef_) # w (ağırlıklar) , Katsayılar (eğim)
print(model.intercept_) # s (sabit terim)
```

```python
# Saçılım grafiği oluştur
plt.scatter(X_test, y_test, color='black') # Gerçek veriler (test verileri) siyah noktalarla gösterilir

# Tahmin edilen değerleri çiz
plt.plot(X_test, y_pred, color='blue', linewidth=3) # Tahmin edilen değerler mavi çizgi ile gösterilir

# Etiketler ve başlık ekle
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progession Against BMI')

# Grafiği göster
plt.show()
```
